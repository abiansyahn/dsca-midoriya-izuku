```python
count_per_age = df_heart.groupby(["age"]).count()

plt.figure(figsize=(10,6))
plt.bar(count_per_age.index.values, count_per_age["target"])
plt.title("Total Heart Disease Patiant per Ages")
plt.xlabel("Age(s)")
plt.ylabel("Total Patient")
plt.show()
```

```python
count_per_age
```

Bar Chart is choosen because since it's the most powerful visualization for compairing categorical data

58 y.o is the age that has the most case of heart disease

```python
age_group = df_heart["age"].unique()
age_group.sort()


avg_rbp = []
for age in age_group:
    current_avg = df_heart.loc[df_heart["age"] == age]["trestbps"].mean()
    avg_rbp.append(current_avg)

plt.plot(age_group, avg_rbp)
plt.title("The Progres of Avg Resting Blood Pressure by Age")
plt.xlabel("Age")
plt.ylabel("Avg. Resting Blood Pressure")
plt.show()
```

Line Chart is choosen because we can se the progress between age clearly, and we can see if its increase or decrease over age.
From the visualization, the 2 age that have the most steepest difference of resting blood pressure is 58-59
